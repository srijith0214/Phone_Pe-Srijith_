# Aggregated Transaction data India

```python
import os
path="/content/pulse/data/aggregated/transaction/country/india/"
Agg_India_list=os.listdir(path)
Agg_India_list.remove('state')
Agg_India_list
```

```python
# Required libraries
import pandas as pd
import json

# Initialize dictionary to store transaction data for India (aggregated at country level)
cli = {
    'Year': [],
    'Quater': [],
    'Transacion_type': [],
    'Transacion_count': [],
    'Transacion_amount': []
}

# Use the list of years from the India-level aggregated data
Agg_yr = Agg_India_list  # Make sure Agg_India_list is defined earlier

# Loop through each year folder
for i in Agg_India_list:
    p_i = path + i + "/"  # Path to the year folder
    Agg_qtr = os.listdir(p_i)  # List of quarter JSON files

    # Loop through each quarter JSON file
    for j in Agg_qtr:
        path_j = p_i + j  # Full path to the JSON file

        # Open and load the JSON data
        with open(path_j, 'r') as Data:
            D = json.load(Data)

            # Extract transaction data
            for z in D['data']['transactionData']:
                Name = z['name']
                count = z['paymentInstruments'][0]['count']
                amount = z['paymentInstruments'][0]['amount']

                # Append extracted data to the dictionary
                cli['Transacion_type'].append(Name)
                cli['Transacion_count'].append(count)
                cli['Transacion_amount'].append(amount)
                cli['Year'].append(i)
                cli['Quater'].append(int(j.strip('.json')))

# Convert the dictionary to a pandas DataFrame
Agg_Ind = pd.DataFrame(cli)
```

# Aggregated User State

```python
import os
import json
import pandas as pd

path = "/content/pulse/data/aggregated/user/country/india/state/"
Agg_user_list = os.listdir(path)

Agg_user_state = {
    'State': [], 'Year': [], 'Quater': [],
    'registeredUsers': [], 'appOpens': []
}

Users_by_devices = {
    'State': [], 'Year': [], 'Quater': [],
    'Brand': [], 'Brand_count': [], 'Brand_percentage': []
}

for i in Agg_user_list:
    p_i = path + i + "/"
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = p_i + j + "/"
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = p_j + k

            try:
                with open(p_k, 'r') as Data:
                    D = json.load(Data)

                    # Extract aggregated data
                    registeredUsers = D['data']['aggregated'].get('registeredUsers')
                    appOpens = D['data']['aggregated'].get('appOpens')

                    Agg_user_state['registeredUsers'].append(registeredUsers)
                    Agg_user_state['appOpens'].append(appOpens)
                    Agg_user_state['State'].append(i)
                    Agg_user_state['Year'].append(j)
                    Agg_user_state['Quater'].append(int(k.strip('.json')))

                    # Extract usersByDevice only if it's a list
                    usersByDevice = D['data'].get('usersByDevice')
                    if isinstance(usersByDevice, list):
                        for device in usersByDevice:
                            Brand = device.get('brand')
                            Brand_count = device.get('count')
                            Brand_percentage = device.get('percentage')

                            Users_by_devices['Brand'].append(Brand)
                            Users_by_devices['Brand_count'].append(Brand_count)
                            Users_by_devices['Brand_percentage'].append(Brand_percentage)
                            Users_by_devices['State'].append(i)
                            Users_by_devices['Year'].append(j)
                            Users_by_devices['Quater'].append(int(k.strip('.json')))

            except Exception as e:
                print(f"Error processing file {p_k}: {e}")

# Convert to DataFrames
Agg_User_State = pd.DataFrame(Agg_user_state)
Users_by_Devices = pd.DataFrame(Users_by_devices)

# Optional: display or save the DataFrames
# print(Agg_User_State)
# print(Users_by_Devices)
```

```python
# Define the path to the aggregated user data for India
path = "/content/pulse/data/aggregated/user/country/india/"

# List all items in the directory and remove 'state' as it's not needed
Agg_India_user = os.listdir(path)
Agg_India_user.remove('state')

# Initialize dictionary to store aggregated user data
Agg_user_Ind = {
    'Year': [],
    'Quater': [],
    'registeredUsers': [],
    'appOpens': []
}

# Initialize dictionary to store user device data
Users_by_devices = {
    'Year': [],
    'Quater': [],
    'Brand': [],
    'Brand_count': [],
    'Brand_percentage': []
}

# Loop through each year folder in the India user data
for i in Agg_India_user:
    p_i = path + i + "/"  # Construct path to the year folder
    Agg_user = os.listdir(p_i)  # List all JSON files (quarters) in the year folder

    # Loop through each quarter JSON file
    for j in Agg_user:
        path_j = p_i + j  # Full path to the JSON file

        try:
            # Open and load the JSON data
            with open(path_j, 'r') as Data:
                D = json.load(Data)

                # Extract registered users and app opens
                registeredUsers = D['data']['aggregated'].get('registeredUsers')
                appOpens = D['data']['aggregated'].get('appOpens')

                # Append data to the aggregated user dictionary
                Agg_user_Ind['registeredUsers'].append(registeredUsers)
                Agg_user_Ind['appOpens'].append(appOpens)
                Agg_user_Ind['Year'].append(i)
                Agg_user_Ind['Quater'].append(int(j.strip('.json')))

                # Extract device usage data if available
                usersByDevice = D['data'].get('usersByDevice')
                if isinstance(usersByDevice, list):
                    for device in usersByDevice:
                        Brand = device.get('brand')
                        Brand_count = device.get('count')
                        Brand_percentage = device.get('percentage')

                        # Append device data to the device dictionary
                        Users_by_devices['Brand'].append(Brand)
                        Users_by_devices['Brand_count'].append(Brand_count)
                        Users_by_devices['Brand_percentage'].append(Brand_percentage)
                        Users_by_devices['Year'].append(i)
                        Users_by_devices['Quater'].append(int(j.strip('.json')))
        except Exception as e:
            # Print error message if file processing fails
            print(f"Error processing file {path_j}: {e}")

# Convert dictionaries to pandas DataFrames for further analysis
Agg_Ind_User = pd.DataFrame(Agg_user_Ind)
Users_by_devices = pd.DataFrame(Users_by_devices)
```
